```python
#B.3 Procedure


# Define the chunk duration in milliseconds
chunk_duration = 30/1000  # You can try different durations and observe the effects

# Number of samples per chunk
samples_per_chunk = int(chunk_duration * sampling_rate )

# Function to segment the audio data into chunks
def segment_audio(audio_data, samples_per_chunk):
    num_chunks = len(audio_data) // samples_per_chunk
    return [audio_data[i * samples_per_chunk:(i + 1) * samples_per_chunk] for i in range(num_chunks)]

# Turn Audio data into consecutive chunks
consecutive_chunks = segment_audio(audio_data, samples_per_chunk)

#turn audio data into overlapping chunks
overlap_chunk_size = samples_per_chunk // 2
overlap_chunks = []
for i in range(0, len(audio_data) - overlap_chunk_size, overlap_chunk_size):
    overlap_chunks.append(audio_data[i:i + samples_per_chunk])

#display number of each chunk
print(f"Number of Consecutive Chunks: {len(consecutive_chunks)}")
print(f"Number of Overlapping Chunks: {len(overlap_chunks)}")


# plot each chunk in different graphs
plt.figure(figsize=(14, 4))
plt.subplot(1, 2, 1)
plt.plot(consecutive_chunks[2])
plt.title("Consecutive Chunk")
plt.subplot(1, 2, 2)
plt.plot(overlap_chunks[2])
plt.title("Overlapping Chunk")
plt.tight_layout()
plt.show()
```

```python
#Frequency-Domain-Analysis 20%

# Define the parameters for the band-pass filters
# Center frequencies for bandpass filters in Hz
base_frequency = list(range(50, 6100, 50))
bandwidth = 50 # Bandwidth of each BPF in Hz

# Function to apply a bank of band-pass filters to a chunk
def bandpass_filters(single_chunk, base_frequency, bandwidth, sampling_rate):
    bands = []
  #bandpass filter using band width and centre freq to set hig and low cutoff
    for center_freq in base_frequency:
        nyquist = sampling_rate / 2
        low = (center_freq - bandwidth / 2) / nyquist
        high = (center_freq + bandwidth / 2) / nyquist
        b, a = butter(4, [low, high], btype='band')

        # Apply the band-pass filter
        filtered = lfilter(b, a, single_chunk)
        bands.append(filtered)

    return bands

# RMS of each chunk
def calculate_rms(single_chunk):
  rms = np.sqrt(np.mean(single_chunk**2))
  return rms

#new list of audio wave
def complete_sine_wave(amplitude,center_frequency,time_domain):
  sine_wave_final= amplitude*np.sin(2*np.pi*center_frequency*time_domain)
  return sine_wave_final

#using bandpass filter on each chunk sepearately
filtered_audio = []
for single_chunk in consecutive_chunks:
  filtered_result = bandpass_filters(single_chunk,base_frequency,bandwidth,sampling_rate)
  filtered_audio.append(filtered_result)

```

```python
#Synthesis 20%

# Create a array for values of the time data for entire duration
time_array = np.linspace(0,len(audio_data)/sampling_rate,len(audio_data))

# Divider time array. Result is time domains relating to each chunk
chunked_time_array = np.array_split(time_array, len(filtered_audio))

#initilaize array for sine wave to input chunks
final_wave = []

#loop through array of filtered audio

for i in range(0,len(filtered_audio)):
  # store specific time chunk into variable chunk
  chunk = filtered_audio[i]
  # access the specific time period corresponding to that time chunk
  current_time_array = np.array(chunked_time_array[i])
  #create array to store chunk audio data
  complete_audio_chunk = np.zeros_like(current_time_array)

  #loop through chunk and look at frequecny

  for j in range(0,len(base_frequency)):
    #calculate amplitude
    amplitude = calculate_rms(chunk[j])

    #making a sine wave by adding frequency and time duration amplitude of each chunk
    completed_wave = complete_sine_wave(amplitude,base_frequency[j],current_time_array)

    #add each chunk to an array to input the data
    complete_audio_chunk = complete_audio_chunk + completed_wave

  #add each chunk to a full sine wave signal
  final_wave.append(complete_audio_chunk)

#making a array without sepearation between chunks
flattened_final_wave = [item for sublist in final_wave for item in sublist]

#changing that array and set 16 bit
audio_samples = np.array(flattened_final_wave,dtype = np.int16)


#making and naming of wav file
wf.write("robotVoice.wav", sampling_rate, audio_samples)

#save data of the roboit voice as sampling rate and audio data
sampling_rate, audio_data = wf.read("robotVoice.wav")



# Plot the audio
plt.figure(figsize=(10, 4))
plt.plot( np.linspace(0., len(audio_data) / sampling_rate, len(audio_data)), audio_data)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Audio Time Stream')
plt.grid(True)
plt.tight_layout()
plt.savefig('audio_time_stream.png')  # Save the plot as an image file
plt.show()
from IPython.display import Audio
Audio(audio_data,rate = sampling_rate)
```
